```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("wheat.csv")
df.head().T
```

```python
yearly_rainfall = df.groupby(['Price_Date']).sum()
yearly_rainfall['rise_fall'] = np.where(yearly_rainfall['Modal_Price'] > yearly_rainfall['Modal_Price'].shift(1), "Rise", "Fall")
year = yearly_rainfall.index
plt.style.use('seaborn-whitegrid')
fig = plt.figure(figsize=(28, 10))

ax = sns.pointplot(x=year, y="Modal_Price", data=yearly_rainfall, color='grey')
ax = sns.pointplot(x=year, y="Modal_Price", data=yearly_rainfall,hue='rise_fall', markers=["x", "o"], join=False)
ax = plt.xticks(rotation=90)
ax = plt.title('Commodity Trend in India')
```

```python
df=pd.read_csv("wheat.csv")
df.head()
```

# K NEAREST NEIGHBOUR ALGORITHM

```python
from sklearn.neighbors import KNeighborsClassifier as KNNClassifier
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df[['Min_Price','Max_Price','Id_No','Sl_No']],df["Modal_Price"],test_size=0.3, random_state=42)
```

```python
knn = KNNClassifier().fit(x_train, y_train)
```

```python
print(knn.score(x_train, y_train))
print(knn.score(x_test, y_test))
print('-'*20)
```

# Final Accuracy of KNN

```python
print('Final Accuracy of KNN :', knn.score(x_test, y_test))
```

# SUPPORT VECTOR MACHINE ALGORITHM

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['Min_Price','Max_Price','Id_No','Sl_No']],df["Modal_Price"],test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
from sklearn.model_selection import train_test_split
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

# Final Accuracy of SVM

```python
print('Test accuracy of sklearn SVM: {}'.format(model.score(X_test, y_test)))
```
